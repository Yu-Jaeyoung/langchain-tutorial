```python
%%time

import pandas as pd
from langchain_core.documents import Document


def load_csv_like_unstructured_csv_loader(file_path: str) -> list[Document]:
    df = pd.read_csv(file_path)

    return [
        Document(
            page_content=df.to_csv(index=False),
            metadata={
                "source": file_path,
                "text_as_html": df.to_html(index=False),
            },
        )
    ]


documents = load_csv_like_unstructured_csv_loader("/workspace/day_02/test.csv")

print("청크의 수:", len(documents))
```
